#### import library dan dataser

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('Drug_clean.csv')
```

#### menampilkan dataset

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
data = df.groupby('Reviews').mean()
data.head()
```

#### menampilkan scatter plot

```python
plt.scatter(df['Reviews'], df['Effective'])
plt.title('Data ke efektifan berdasarkan reviews')
plt.xlabel('Reviews')
plt.ylabel('Effective')
plt.show()
```

#### menampilkan pie chart

```python
data = df.Form.value_counts()
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.title('Persentase Form')
plt.show()
```

#### menampilkan heat map

```python
cor_matrix = df.corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.show()
```
